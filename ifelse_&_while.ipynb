```python
correct_email = 'xyz@gmail.com'
correct_password = '1234'

email = input("enter email : ")
if '@' in email:
  password = input("enter password : ")


if email == correct_email and password == correct_password:
  print("welcome")
elif email == correct_email and password != correct_password:
  print("please enter valid password")
  password_again = input("enter password : ")
  if password_again == correct_password:
    print('login successful')
  else:
    print('still incorrect')

else:
  print("invalid email")

```

```python
import random
jackpot = random.randrange(1,50)
# print(jackpot)
counter = 1
guess_number = int(input("guess the number : "))

while guess_number != jackpot:
  if guess_number < jackpot:
    print("guess higher")
  else:
    print("guess lower")
  guess_number = int(input("try again : "))
  counter += 1

print(f'you guessed correctly in {counter} attempts')
```
